##Generate Train & Test Dataset

**Copy** the code and **run locally** to generate train and test dataset


```
import os
full_path_to_images = ''
colab_path = ""
#Rename Colab path to data/ the name of your folder with image and 
#annotations
os.chdir(full_path_to_images)
p = []
for current_dir, dirs, files in os.walk('.'):
    for f in files:
        if f.endswith('.jpg'):
            # Preparing path to save into train.txt file
            # If you're using Windows, it might need to change
            # this: + '/' +
            # to this: + '\' +
            # or to this: + '\\' +
            path_to_save_into_txt_files = colab_path + '/' + f
            p.append(path_to_save_into_txt_files + '\n')
p_test = p[:int(len(p) * 0.15)]
p = p[int(len(p) * 0.15):]
with open('train.txt', 'w') as train_txt:
    for e in p:
        train_txt.write(e)
with open('test.txt', 'w') as test_txt:
    for e in p_test:
        test_txt.write(e)
print('done')
```

**Fill** up the full_path_to_images and colab_path with respective path values
